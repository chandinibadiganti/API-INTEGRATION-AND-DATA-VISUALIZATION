## Import Required Libraries

The required Python libraries are imported for API requests,
data handling, and visualization.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Fetch Data From Public API

Using Open-Meteo API, weather data is collected without an API key.

```python
url = "https://api.open-meteo.com/v1/forecast?latitude=12.9716&longitude=77.5946&current_weather=true"
response = requests.get(url)
data = response.json()
data
```

## Data Extraction

The required weather parameters are extracted from the API response.

```python
weather = {
    "Temperature": data["current_weather"]["temperature"],
    "Wind Speed": data["current_weather"]["windspeed"],
    "Weather Code": data["current_weather"]["weathercode"]
}

weather
```

## Convert Data Into DataFrame

The extracted data is converted into a pandas dataframe for analysis.

```python
df = pd.DataFrame([weather])

df
```

## Data Visualization Dashboard

A bar chart is created using Matplotlib and Seaborn
to visualize weather parameters.

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=df.columns,
    y=df.iloc[0]
)

plt.title("Weather Data Visualization Dashboard")
plt.xlabel("Parameters")
plt.ylabel("Values")

plt.show()
```

## Save Output Files

The dataset and dashboard image are saved for submission.

```python
df.to_csv("weather_data.csv", index=False)

plt.figure(figsize=(8,5))

sns.barplot(
    x=df.columns,
    y=df.iloc[0]
)

plt.title("Weather Data Visualization Dashboard")

plt.savefig("dashboard.png")

plt.show()
```

## Weather Data Visualization Dashboard

Multiple charts are created to analyze different weather parameters.


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,5))

# Temperature chart
plt.subplot(1,3,1)
sns.barplot(
    x=["Temperature"],
    y=[df["Temperature"][0]]
)
plt.title("Temperature")

# Wind Speed chart
plt.subplot(1,3,2)
sns.barplot(
    x=["Wind Speed"],
    y=[df["Wind Speed"][0]]
)
plt.title("Wind Speed")

# Weather Code chart
plt.subplot(1,3,3)
sns.barplot(
    x=["Weather Code"],
    y=[df["Weather Code"][0]]
)
plt.title("Weather Condition")

plt.tight_layout()

plt.savefig("dashboard.png")

plt.show()
```
